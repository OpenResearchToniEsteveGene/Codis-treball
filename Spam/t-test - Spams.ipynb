```python
# Import libraries for numerical computation, statistics, and plotting

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd  # ← Required to create the final table

# Actual AUC-ROC values for two models: TF-IDF and LSA
# (each list contains the results for 5 cross-validation folds)
auc_tfidf = [
    0.9806632124352331, 
    0.9831778929188255, 
    0.9831206314984179, 
    0.9789825086066001, 
    0.9868484195152484
]

auc_lsa = [
    0.9878916437736899, 
    0.9658552700212122, 
    0.9892756546232223, 
    0.9810224525043177, 
    0.9848497409326425
]

# Compute the pairwise differences between LSA and TF-IDF
# (to observe which model performs better in each fold)

differences = [a - b for a, b in zip(auc_lsa, auc_tfidf)]

# Q-Q Plot: graphical check of the normality of the differences
# This plot compares the observed quantiles with those of a theoretical normal distribution

plt.figure(figsize=(6, 6))
stats.probplot(differences, dist="norm", plot=plt)
plt.title("Q-Q Plot of the differences (LSA - TF-IDF)")
plt.xlabel("Theoretical quantiles")
plt.ylabel("Observed quantiles")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Compute the paired Student's t-test (ttest_rel)
# Objective: to check whether the mean difference between LSA and TF-IDF is statistically significant

from scipy.stats import ttest_rel


# Apply the paired t-test to compare LSA vs TF-IDF
t_statistic_real, p_value_real = ttest_rel(auc_lsa, auc_tfidf)

# Create a comparison table with the metrics and differences

df_real = pd.DataFrame({
    "Fold": list(range(1, 6)),  # Fold numbers
    "AUC_TFIDF": auc_tfidf,  # TF-IDF results
    "AUC_LSA": auc_lsa,      # LSA results
    "Difference (LSA - TF-IDF)": [a - b for a, b in zip(auc_lsa, auc_tfidf)]
})

# Final result of the t-test
# t_statistic_real: t-value
# p_value_real: associated p-value (used to decide whether the difference is significant)

t_statistic_real, p_value_real
```
